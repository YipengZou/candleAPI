```python
import ipython_importer
from _loadData import loadData
import talib
from pyecharts import options as opts
from pyecharts.charts import Kline, Line, Bar, Grid, Tab
import numpy as np
from pyecharts.commons.utils import JsCode
tab = Tab()
def main(code):
    for freq in ["600S", "300S", "180S", "60S"]:
        test1 = loadData(code, freq)
        data_use = test1.df_concat
        data = np.array(data_use[["open", "close", "low", "high"]]).tolist()
        kline = (
            Kline(init_opts=opts.InitOpts(
                    animation_opts=opts.AnimationOpts(animation_easing="linear")))
            .add_xaxis([str(i) for i in data_use.index])
            .add_yaxis(
                "Stock Info",
                data,
                itemstyle_opts=opts.ItemStyleOpts(
                    color="#ef232a",
                    color0="#14b143",
                    border_color="#ef232a",
                    border_color0="#14b143",
                ),
            )
            .set_global_opts(
                title_opts=opts.TitleOpts(title=code + " " + freq + " ", pos_left="0"),
                xaxis_opts=opts.AxisOpts(
                    type_="category",
                    is_scale=True,
                    boundary_gap=False,
                    axisline_opts=opts.AxisLineOpts(is_on_zero=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False),
                    split_number=20,
                    min_="dataMin",
                    max_="dataMax",
                ),
                yaxis_opts=opts.AxisOpts(
                    is_scale=True, splitline_opts=opts.SplitLineOpts(is_show=True)
                ),
                tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="line"),
                datazoom_opts=[
                    opts.DataZoomOpts(
                        is_show=False, type_="slider", xaxis_index=[0, 0], range_end=100
                    ),
                    opts.DataZoomOpts(
                        is_show=True, xaxis_index=[0, 1], pos_top="97%", range_end=100
                    ),
                    opts.DataZoomOpts(is_show=False, xaxis_index=[0, 2], range_end=100),
                ],
            )
        )
        kline_line = (
            Line()
            .add_xaxis([str(i) for i in data_use.index])
            .add_yaxis(
                series_name="MA5",
                y_axis=data_use["MA5"],
                is_smooth=True,
                linestyle_opts=opts.LineStyleOpts(opacity=0.5),
                label_opts=opts.LabelOpts(is_show=False),
            )
            .add_yaxis(
                series_name="MA10",
                y_axis=data_use["MA10"],
                is_smooth=True,
                linestyle_opts=opts.LineStyleOpts(opacity=0.5),
                label_opts=opts.LabelOpts(is_show=False),
            )
            .add_yaxis(
                series_name="MA20",
                y_axis=data_use["MA20"],
                is_smooth=True,
                linestyle_opts=opts.LineStyleOpts(opacity=0.5),
                label_opts=opts.LabelOpts(is_show=False),
            )
            .add_yaxis(
                series_name="MA30",
                y_axis=data_use["MA30"],
                is_smooth=True,
                linestyle_opts=opts.LineStyleOpts(opacity=0.5),
                label_opts=opts.LabelOpts(is_show=False),
            )
            .set_global_opts(
                xaxis_opts=opts.AxisOpts(
                    type_="category",
                    grid_index=1,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                ),
                yaxis_opts=opts.AxisOpts(
                    grid_index=1,
                    split_number=3,
                    axisline_opts=opts.AxisLineOpts(is_on_zero=False),
                    axistick_opts=opts.AxisTickOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False),
                    axislabel_opts=opts.LabelOpts(is_show=True),
                ),
            )
        )
        overlap_kline_line = kline.overlap(kline_line)

        bar_1 = (
            Bar()
            .add_xaxis([str(i) for i in data_use.index])
            .add_yaxis(
                series_name="Volume",
                y_axis=data_use["volume"].to_list(),
                xaxis_index=1,
                yaxis_index=1,
                label_opts=opts.LabelOpts(is_show=False),
                itemstyle_opts=opts.ItemStyleOpts(
                    color=JsCode(
                        """
                        function(params) {
                        var colorList;
                        if (barData[params.dataIndex][1] >= barData[params.dataIndex][0]) {
                            colorList = '#ef232a';
                        } else {
                            colorList = '#14b143';
                        }
                        return colorList;
                    }
                    """
                    )
                ),
            )
            .set_global_opts(
                xaxis_opts=opts.AxisOpts(
                    type_="category",
                    grid_index=1,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                ),
                legend_opts=opts.LegendOpts(is_show=False),
            )
        )
        bar_2 = (
            Bar()
            .add_xaxis([str(i) for i in data_use.index])
            .add_yaxis(
                series_name="MACD",
                y_axis=data_use["MACD"].to_list(),
                xaxis_index=2,
                yaxis_index=2,
                label_opts=opts.LabelOpts(is_show=False),
                itemstyle_opts=opts.ItemStyleOpts(
                    color=JsCode(
                        """
                            function(params) {
                                var colorList;
                                if (params.data >= 0) {
                                colorList = '#ef232a';
                                } else {
                                colorList = '#14b143';
                                }
                                return colorList;
                            }
                            """
                    )
                ),
            )
            .set_global_opts(
                xaxis_opts=opts.AxisOpts(
                    type_="category",
                    grid_index=2,
                    axislabel_opts=opts.LabelOpts(is_show=False),
                ),
                yaxis_opts=opts.AxisOpts(
                    grid_index=2,
                    split_number=4,
                    axisline_opts=opts.AxisLineOpts(is_on_zero=False),
                    axistick_opts=opts.AxisTickOpts(is_show=False),
                    splitline_opts=opts.SplitLineOpts(is_show=False),
                    axislabel_opts=opts.LabelOpts(is_show=True),
                ),
                legend_opts=opts.LegendOpts(is_show=False),
            )
        )   

        line_2 = (
            Line()
            .add_xaxis([str(i) for i in data_use.index])
            .add_yaxis(
                series_name="DEA",
                y_axis=data_use["DEA"],
                xaxis_index=2,
                yaxis_index=2,
                label_opts=opts.LabelOpts(is_show=False),
            )
            .add_yaxis(
                series_name="DIF",
                y_axis=data_use["DIF"],
                xaxis_index=2,
                yaxis_index=2,
                label_opts=opts.LabelOpts(is_show=False),
            )
            .set_global_opts(legend_opts=opts.LegendOpts(is_show=False))
        )
        # 
        overlap_bar_line = bar_2.overlap(line_2)

        grid_chart = Grid(init_opts=opts.InitOpts(width="1000px", height="800px"))
        grid_chart.add_js_funcs("var barData = {}".format(np.array(data_use[["open", "close"]]).tolist()))
        grid_chart.add(
            overlap_kline_line,
            grid_opts=opts.GridOpts(pos_left="7%", pos_right="1%", height="60%"),
        )
        grid_chart.add(
            bar_1,
            grid_opts=opts.GridOpts(pos_left="7%", pos_right="1%",pos_top="71%", height="10%"),
        )
        grid_chart.add(
            overlap_bar_line,
            grid_opts=opts.GridOpts(
                pos_left="7%", pos_right="1%", pos_top="82%", height="14%",),
        )
        tab.add(grid_chart, freq)
    tab.render_notebook()
    return tab, bar_1
stockCode = input("(000001.SZ): ")
tab, bar_1 = main(stockCode)
tab.render("../demo/echarts-" + stockCode + " .html")
tab.render_notebook()
```
